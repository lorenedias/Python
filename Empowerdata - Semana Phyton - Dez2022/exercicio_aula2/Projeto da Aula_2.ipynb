```python
# importação das bibliotecas 
import yfinance
import pyautogui
import pyperclip
import time
```

### Passo 1 - Obtendo os dados Yahoo Finanças

```python
# definindo qual a ação será analisada
ticker = input("Digite o código da ação: ")
destinatario = input("Digite o e-mail do destinatário: ")
```

```python
dados = yfinance.Ticker(ticker)
```

```python
# analisando os dados históriocos da ação escolhida
dados.history()
```

```python
# filtrando os dados dos últimos seis meses da ação escolhida
tabela = dados.history("6mo")
tabela
```

```python
# filtrando os dados dos últimos seis meses da ação escolhida, apenas a coluna que traz os valores de fechamento da ação
fechamento = tabela.Close
fechamento
```

```python
# gerando um gráfico simples para análise
grafico = fechamento.plot()
```

### Passo 2 - obtendo as medidas solicitadas

```python
#gerando os dados estáticos solicitados (máximo, mínimo e atual)

maximo = fechamento.max()
minimo = fechamento.min()
atual = fechamento[-1]
```

```python
#imprimindo os resultados na tela
print(maximo)
print(minimo)
print(atual)
```

###  Passo 3 - enviando o e-mail automáticamente

```python
#configurar uma pausa entre as ações do pyautogui
pyautogui.PAUSE = 3
```

```python
#acessando a uma nova aba e o site do Gmail
pyautogui.hotkey("ctrl","t")

pyperclip.copy("www.gmail.com")
pyautogui.hotkey("ctrl","v")
pyautogui.hotkey("enter")
```

```python
#acessando a o botão escrever por atalho , para utilizar é necessário alterar a configuração do Gmail: 
#Configurações / seção: Atalhos do teclado / opção: Atalhos do teclado ativados
pyautogui.hotkey("c")
```

```python
#preenchendo o destinatário
pyperclip.copy(destinatario)

pyautogui.hotkey("ctrl","v")
pyautogui.hotkey("tab")
```

```python
#preenchendo o assunto
assunto = f"Análise Diária {ticker}"

pyperclip.copy(assunto)
pyautogui.hotkey("ctrl","v")
pyautogui.hotkey("tab")
```

```python
#preenchendo a mensagem
mensagem = f"""
Bom dia!

Segue abaixo as análises da ação {ticker} dos últimos seis meses:

Cotação máxima: R$ {round(maximo,2)}
Cotação mínima: R$ {round(minimo,2)}
Cotação atual: R$ {round(atual,2)}

Atendiosamente,
XXXXXXXXX XXXXXXXX
"""
pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl","v")
```

```python
#enviando o e-mail
pyautogui.hotkey("ctrl","enter")
```
